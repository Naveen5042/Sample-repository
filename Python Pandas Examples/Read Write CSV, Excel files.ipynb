```python
import pandas as pd
df = pd.read_csv("zoo.csv",header=None,names=['animal','uniq_id','water_need'],nrows=2)
df
#If you dont have headers in ur csv file then names[] is used to add headers to ur DataFrame
#nrows is used to get particular number of rows as your DataFrame from your csv file
```

```python
df=pd.read_csv("zoo.csv",na_values=["not available","n.a"])
df
#Here you are saying like if you find any of the value in [not available,n.a] then replace those to Nan(Not a Number)
#If you want to replace only one column with Nan then refer the below cell
```

```python
df = pd.read_csv("zoo.csv",na_values={
    "animal":['not available','n.a'],
    'uniq_id':['not available','n.a',-1]
})
df
#This is something like customizing. -1 will get replaced with Nan only for uniq_id column
```

```python
#How to replace the content of field with some thing using converter
def convert_product(cell):
    if(cell=="n.a"):
        return "Carretera"
    else:
        return cell
def convert_discount(cell):
    if(cell==0):
        return None
    else:
        return cell
df = pd.read_excel("Financial Sample.xlsx",converters={"Product":convert_product,"Discounts":convert_discount})
df.to_excel("new_excel.xlsx",startrow=2,startcol=2,index=False)
#When ever you are reading Product column, call convert_product method for every field. 
#Converters will take dictionary(key-->Columns, value="name of the method")
#You can do for multiple columns also like as shown in above
#You can start writing from particular row and column as shown above example
```
